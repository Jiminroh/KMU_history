# Stemming(어간 추출), Lemmatizing(원형 복원), 품사 부착(POS-tagging)

## Stemming
- <span style = 'font-size:1.3em;line-height:1.5em'>단어의 접미사나 어미를 제거하여 같은 의미를 갖는 형태소의 기본형을 찾는 방법</span>
    - <span style = 'font-size:1.3em;line-height:1.5em'>Stemming은 단순히 어미를 제거할 뿐, 정확한 원형을 찾아주지는 않음</span>

```python
from nltk.stem import PorterStemmer, LancasterStemmer
stemmer1 = PorterStemmer()
stemmer2 = LancasterStemmer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']
print(f"Porter stemmer result : {[stemmer1.stem(w) for w in words]}")
print(f"LancasterStemmer stemmer result : {[stemmer2.stem(w) for w in words]}")
print()

words = ['go', 'went', 'gone', 'going']
print(f"Porter stemmer result : {[stemmer1.stem(w) for w in words]}")
print(f"LancasterStemmer stemmer result : {[stemmer2.stem(w) for w in words]}")
```

## Lemmatizing
- <span style = 'font-size:1.3em;line-height:1.5em'>같은 의미를 갖는 여러 단어를 사전형으로 통일하는 작업</span>
    - <span style = 'font-size:1.3em;line-height:1.5em'>POS tagging을 달아줄 경우 더 정확한 원형을 찾을 수 있음</span>

```python
# nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']
print(f"Lemmatizer result : {[lm.lemmatize(w, pos='v') for w in words]}")
words = ['go', 'went', 'gone', 'going']
print(f"Lemmatizer result : {[lm.lemmatize(w, pos='v') for w in words]}")
```

## POS-tagging
- <span style = 'font-size:1.3em;line-height:1.5em'>낱말을 문법적인 기능이나 형태, 뜻에 따라 구분한 것</span>
- <span style = 'font-size:1.3em;line-height:1.5em'>NLTK에선 Penn Treebank Tagset을 활용. 다음은 품사 tagging 예시</span>
    - <span style = 'font-size:1.1em;line-height:1.5em'>NNP: 단수 고유명사</span>
    - <span style = 'font-size:1.1em;line-height:1.5em'>VB: 동사</span>
    - <span style = 'font-size:1.1em;line-height:1.5em'>VBP: 동사 현재형</span>
    - <span style = 'font-size:1.1em;line-height:1.5em'>TO: to 전치사</span>
    - <span style = 'font-size:1.1em;line-height:1.5em'>NN: 명사</span>

# FreqDist

- <span style = 'font-size:1.3em;line-height:1.5em'>문서에 사용된 단어 (혹은 토큰)의 사용빈도 정보를 담는 클래스</span>

```python
from nltk import FreqDist
```

```python
my_text
```

```python
all_tokens = reg_tokenizer.tokenize(my_text)
stop_words_list = stopwords.words('english')
print(len(all_tokens))
print(len(stop_words_list))
```

```python
all_tokens = [token for token in all_tokens if token not in stop_words_list]
print(len(all_tokens))
```

```python
all_tokens_pos = pos_tag(all_tokens)
print(all_tokens_pos[:20])
```

### NNP(고유명사) 의 빈도수 구하기

```python
NNP_tokens_pos = [token for token in all_tokens_pos if token[1]=='NNP']
len(NNP_tokens_pos)
```

```python
NNP_tokens_pos[:30]
```

```python
stopwords1 = [token for token in NNP_tokens_pos if token[0].startswith('_')]
stopwords2 = ['Mr','Mrs','Miss','Mr.','Mrs.', 'Dear','Ms']
NNP_tokens_pos = [token for token in NNP_tokens_pos if token[0] not in stopwords1+stopwords2]
NNP_tokens = [token[0] for token in NNP_tokens_pos]
```

```python
len(NNP_tokens_pos)
```

```python
NNP_tokens_pos
```

```python
fd = FreqDist(NNP_tokens)
```

```python
#### 전체 NNP단어 수
fd.N()
```

```python
#### 'Emma'가 나온 빈도수
fd['Emma']
```

```python
#### 'Emma'가 나올 확률
fd.freq('Emma')
```

```python
#### 가장 자주 발생한 단어들(NNP)
fd.most_common(10)
```

# Assignment

<span style = 'font-size:1.3em;line-height:1.5em'><b>1.</b> nltk의 corpus.gutenberg에 있는 파일들 중 하나를 골라서 전체 preprocessing을 진행해보자.</span>

```python
import nltk.book as book
import nltk.corpus as corpus

# # check files
# print(corpus.gutenberg.fileids())

# # select austen-persuasion.txt
# print(corpus.gutenberg.raw('austen-persuasion.txt'))

# tokenize 
my_text = corpus.gutenberg.raw('austen-persuasion.txt')

# 글을 문장으로 tokenize
my_sentences = sent_tokenize(my_text[50:1000])

# 정규표현식으로 tokenize
reg_tokenizer = RegexpTokenizer("[\w]+")
reg_tokenizer.tokenize(my_sentences[1])

# steaming(어간 추출)
from nltk.stem import PorterStemmer, LancasterStemmer
stemmer1 = PorterStemmer()
stemmer2 = LancasterStemmer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']
print(f"Porter stemmer result : {[stemmer1.stem(w) for w in words]}")
print(f"LancasterStemmer stemmer result : {[stemmer2.stem(w) for w in words]}")
print()

words = ['go', 'went', 'gone', 'going']
print(f"Porter stemmer result : {[stemmer1.stem(w) for w in words]}")
print(f"LancasterStemmer stemmer result : {[stemmer2.stem(w) for w in words]}")

# Lemmatizing(원형 복원)
lm = WordNetLemmatizer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']
print(f"Lemmatizer result : {[lm.lemmatize(w, pos='v') for w in words]}")
words = ['go', 'went', 'gone', 'going']
print(f"Lemmatizer result : {[lm.lemmatize(w, pos='v') for w in words]}")

# pos-tagging(품사 부착)
from nltk.tag import pos_tag, untag
my_sentence = my_sentences[0]
# tagged_list = pos_tag(reg_tokenizer.tokenize(my_sentence))
tagged_list = pos_tag(word_tokenize(my_sentence))

nouns_list = [x[0] for x in tagged_list if x[1] == 'NN']
print(nouns_list)

# Stop Word Elimination
# nltk에 등록된 stopword list확인하기
stop_words_list = stopwords.words('english')
example = my_sentences[1]
stop_words = set(stopwords.words('english'))
word_tokens = reg_tokenizer.tokenize(example)

result = []
for word in word_tokens:
    if word not in stop_words:
        result.append(word)
        
print(f"Before remove stopwords: {word_tokens}")
print(f"After remove stopwords: {result}")

# text 클래스
text = Text(reg_tokenizer.tokenize(my_text))
```
