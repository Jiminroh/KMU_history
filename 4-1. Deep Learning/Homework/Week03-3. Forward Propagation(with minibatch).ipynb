## 이제 MNIST 데이터를 mini-batch (size=100)단위로 forward propagation해보겠습니다.

```python
# mnist.py에서 우리가 실제로 사용할 load_mnist함수만 불러올게요
from mnist import load_mnist
```

```python
(x_trn, y_trn), (x_tst, y_tst) = load_mnist(flatten=True, normalize=False)
```

```python
batch_size = 100
```

```python
idxs=np.arange(0,batch_size).tolist()
```

```python
x_sample = x_trn[idxs]
y_sample = y_trn[idxs]
```

```python
from act_func import *
```

```python
W1, W2, W3 = network['W1'], network['W2'], network['W3']
b1, b2, b3 = network['b1'], network['b2'], network['b3']

a1 = np.dot(x_sample, W1)+b1
z1 = sigmoid(a1)
a2 = np.dot(z1, W2)+b2
z2 = sigmoid(a2)
a3 = np.dot(z2,W3)+b3
a3 = softmax(a3)
```

## 자, 그럼 이제 60000개의 데이터에 대해서 NN의 모든 output을 계산해보죠

```python
result = []
for i in range(0,x_trn.shape[0],batch_size):
    start_idx = i
    end_idx = i+batch_size
    x_sample = x_trn[start_idx:end_idx]
    a1 = np.dot(x_sample, W1)+b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2)+b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3)+b3
    a3 = softmax(a3)
    
    result.extend(np.argmax(a3, axis=1))
result = np.array(result)
```

```python
np.sum(result == y_trn)
```

```python
print(f'미리 학습한 모델의 정확도는 {100*np.sum(result==y_trn) / len(y_trn):.2f}%입니다.')
```

# mini_batch로 계산하는게 single data단위로 계산하는 것보다 빠릅니다
## 시간을 비교해볼까요?

### single data단위

```python
%%time

result1 = []
for i in range(x_trn.shape[0]):
    x_sample = x_trn[i]
    a1 = np.dot(x_sample, W1)+b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2)+b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3)+b3
    a3 = softmax(a3)
    
    result1.append(np.argmax(a3))
result1 = np.array(result1)
```

```python
%%time

result2 = []
for i in range(0,x_trn.shape[0],batch_size):
    start_idx = i
    end_idx = i+batch_size
    x_sample = x_trn[start_idx:end_idx]
    a1 = np.dot(x_sample, W1)+b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2)+b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3)+b3
    a3 = softmax(a3)
    
    result2.extend(np.argmax(a3, axis=1))
result2 = np.array(result2)
```
